```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### GRAPH

```python
x = [0,1,2,3,4,5]
y=[0,2,4,6,8,10]

#resize graph
plt.figure(figsize=(5,3), dpi=100)

#plt.plot(x, y,label='blah',color = 'brown',linewidth=2, linestyle = '--', marker = '*', markersize = 10)

##SHORTHAND fmt = [color][marker][line]
plt.plot(x, y, 'r*--',label='blah')


##Line Number two
#select interval we want to plot our points at
x2 = np.arange(0,4.5,0.5)

#plot part of graph as a line
plt.plot(x2[:6], x2[:6]**2, 'b', label = 'x^2')

#plot remainder of parts as a line
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.title('Our first graph',fontdict={'fontname': 'Cambria', 'fontsize':16})
plt.xlabel('X axis', fontdict = {'fontname' : 'Arial'})
plt.ylabel('Y Axis')

#plt.xticks([1,2,3,4,5])  #keys in our axes, graph resizes instantly
#plt.yticks([2,4,6,8,10])

plt.legend()

##save graph
plt.savefig('mygraph_matplotlib.png', dpi = 300)

plt.show()
```

## Real World Examples

# Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize = (8,5))

plt.title('Gas prices comparison over time in $')

plt.plot(gas.Year, gas.USA,'b.-', label = 'United States')
plt.plot(gas.Year, gas.Canada, label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], label = 'South Korea')

plt.xlabel('Year')
plt.ylabel("US Dollars")

plt.legend()

plt.xticks(gas.Year[::2]) #Iterate every three years, start at beginning, end at end, skip 2 values

plt.show()
```

```python
plt.figure(figsize = (12,8))
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker = '.', label = country)

plt.xlabel('Year')
plt.ylabel("US Dollars")

plt.legend(loc = 'upper left')

plt.xticks(gas.Year[::2])

plt.savefig('Gas Prices.png', dpi = 250)
plt.show()
```

### FIFA DATA

```python
fifa = pd.read_csv('file_data.csv')
fifa
```

### Histogram

```python
#Set bin for interval of 10

bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = '#abcdef')

plt.xticks(bins)

plt.xlabel('Skill level')
plt.ylabel('Number of players')
plt.title('Distribution of Player skills')

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#defabc']
plt.pie([left,right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Preferrence of football player\'s foot')

plt.show()
```

```python
#Converting weight of players to integers from a string
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

#colors
plt.style.use('ggplot')

plt.figure(figsize = (7,7))

light = fifa.loc[fifa.Weight<125].count()[0]
medium_light = fifa.loc[(fifa.Weight>=125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight>150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight>200)].count()[0]

weight = [light, medium_light, medium, medium_heavy, heavy]
labels = ['Under 125','125-150', '150-175', '175-200', 'Higher than 200']

#pctdistance is percentage distance to prevent cluttering
# explode breaks the graph a bit

explode = (.4, .1, 0, 0, .4)

plt.pie(weight, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode )

plt.title('Weight distribution of players')

plt.show()
```

## Lets find out relative strength of clubs (Box and Whisker Plot)

```python
plt.style.use('default')
plt.figure(figsize = (5,10))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
r_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, r_madrid, revs], labels = labels)

colors = ['#e036be', '#2b4ce0', '#c7453c']
for box in boxes['boxes']:
    box.set(color='#e036be')

plt.title('Professoinal football team comparison')
plt.ylabel('Fifa Overall Rating')

plt.show()
```
