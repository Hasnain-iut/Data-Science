```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#train.set_index('Id',inplace=True)
#test.set_index('Id',inplace=True)
```

```python
data = pd.concat([train,test])
```

```python
data['LotFrontage'].isnull().sum()
```

```python
data['LotFrontage'].fillna(0,inplace=True)
```

```python
d = {}
for i in data.columns:
    d[i] = data[i].isnull().sum()
d
```

```python
valid_null_values = data[['BsmtQual','BsmtCond','BsmtExposure', 'BsmtFinType1', 'BsmtFinType2' ,'GarageType',
                          'GarageFinish','GarageQual', 'Alley',
                         'GarageCond','PoolQC','Fence','MiscFeature']]

valid_mean_values = data[['MasVnrArea','BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF',
                          'GarageYrBlt','GarageCars','GarageArea']]

valid_top_values = data[['MSZoning','Utilities','Electrical', 'BsmtFullBath', 
                         'BsmtHalfBath', 'KitchenQual', 'Functional',
                         'FireplaceQu','SaleType',
                        'Exterior1st','Exterior2nd','MasVnrType' ]]
```

```python
for i in valid_null_values.columns:
    data[i].fillna('Null', inplace=True)
    print(i, data[i].isnull().sum())

for j in valid_mean_values.columns:
    mean_val = data[j].mean()
    mean_val = np.ceil(mean_val)
    data[j].fillna(mean_val, inplace=True)
    print(j, data[j].isnull().sum())
    
for k in valid_top_values.columns:
    data[k].fillna(data[k].mode().iloc[0], inplace=True)
    print(k, data[k].isnull().sum())
```

```python
d = {}
for i in data.columns:
    d[i] = data[i].isnull().sum()
d
```

```python
print(data['BsmtFullBath'].value_counts())
print(data['BsmtHalfBath'].value_counts())
print(data['FullBath'].value_counts())
print(data['HalfBath'].value_counts())
```

```python
training = data[data.index.isin(train.index)]
testing = data[data.index.isin(test.index)]
```

```python
categorical_data = training[['MSSubClass', 'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
                         'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual',
                         'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond',
                         'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
                         'HeatingQC', 'CentralAir', 'Electrical', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath',
                         'HalfBath', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
                         'GarageCond','PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']]

numerical_data = training[['LotFrontage', 'LotArea','MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF', 'TotalBsmtSF',
                       '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea','TotRmsAbvGrd', 'Fireplaces',
                       'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch',
                       'PoolArea', 'MiscVal']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for i in numerical_data.columns:
    #plt.scatter(training[i], training['SalePrice'])
    sns.lmplot(x=i, y='SalePrice', data=training)
    plt.title(i)
    plt.show()
```

```python
for i in numerical_data.columns:
    data[i] = data[i].astype(float)

from sklearn.preprocessing import MinMaxScaler

num = data[['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF',
           '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'ScreenPorch']]

minmax = MinMaxScaler()

for i in num.columns:
    data[i] = minmax.fit_transform(data[i].values.reshape(-1,1))
```

```python
data.drop(['MiscVal', 'PoolArea', 'EnclosedPorch','LowQualFinSF','BsmtFinSF2'], axis=1, inplace=True)
```

```python
for i in categorical_data.columns:
    sns.catplot(x = i, y = 'SalePrice', kind='bar', data = training)
```

```python

for i in categorical_data.columns:
    if (data[i].dtype == object):
        data[i] = data[i].astype('category')
        
cat_columns = data.select_dtypes(['category']).columns
data[cat_columns] = data[cat_columns].apply(lambda x: x.cat.codes)
```

```python
trainR = data[data['Id'] <=1460]
testR = data[data['Id'] >1460]
testR.pop('SalePrice')
```

```python
testR
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
X = trainR.drop(['SalePrice', 'Id'], axis=1)
Y = trainR['SalePrice']
x_train, x_test,y_train, y_test = train_test_split(X,Y,  test_size=0.25, random_state=42)

linR = linear_model.LinearRegression()
linR.fit(x_train, y_train)

linR_pred = linR.predict(x_test)

print('Prediction: ', linR_pred)
print('R^2 value: ', linR.score(X,Y))
```

```python
y = testR.drop('Id', axis=1)
testR['SalePrice'] = linR.predict(y)
print(np.array(testR['SalePrice']))
```
