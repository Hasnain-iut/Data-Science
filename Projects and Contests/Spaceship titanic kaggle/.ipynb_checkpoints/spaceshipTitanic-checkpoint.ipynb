## Data Cleaning


```python
data
```

```python
mode_values = data[['HomePlanet', 'CryoSleep', 'Destination']]

for i in mode_values.columns:
    data[i].fillna(data[i].mode().iloc[0], inplace=True)
    print(i, data[i].isnull().sum())
```

```python
data['PassengerGroup'] = data['PassengerId'].str.split('_').str[0]
```

```python
data['VIP'].fillna(False,inplace=True)
```

```python
zero_values = data[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]

for k in zero_values.columns:
    data[k].fillna(0, inplace=True)
    print(k, data[k].isnull().sum())
```

```python
data['Age'].fillna(int(data['Age'].mean()), inplace=True)
```

```python
data[49:54]
```

```python
data.isnull().sum()
```

```python
data.drop('Name', axis=1, inplace=True)
```

## Feature Engineering

```python
data['PassengerGroup'] = data['PassengerId'].str.split('_').str[0]
```

### More cleaning

```python
decks = data[['deck', 'deck_num', 'deck_side']]

for j in decks.columns:
    data[j].fillna(data[j].mode().iloc[0], inplace=True)
    print(j, data[j].isnull().sum())
```
